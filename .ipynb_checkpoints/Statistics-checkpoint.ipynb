```python
class DisplayState:
    def __init__(self, state):
        self.size = len(state)//6
        self.h = int(math.sqrt(self.size))
        self.w = int(math.sqrt(self.size))
        self.raw_state = state
        self.board = state[:self.size].values.reshape(self.h,self.w)
        self.human_trace = state[self.size:2*self.size].values.reshape(self.h,self.w)
        self.computer_trace = state[2*self.size:3*self.size].values.reshape(self.h,self.w)
        self.human_awards = state[3*self.size:4*self.size].values.reshape(self.h,self.w)
        self.computer_awards = state[4*self.size:5*self.size].values.reshape(self.h,self.w)
        self.all_awards = state[5*self.size:6*self.size].values.reshape(self.h,self.w)
        self.dict = {
                    "Board": self.board,
                    "Human trace": self.human_trace,
                    "Computer trace": self.computer_trace,
                    "Human awards": self.human_awards,
                    "Computer awards": self.computer_awards,
                    "All awards": self.all_awards,
                    }
    def ToGrayScale(self, which='all'):
        if(which == 'all'):
            axes=[]
            fig=plt.figure(figsize=(10,8))
            i=0
            j=0
            for key in self.dict:
                axes.append(fig.add_subplot(2, 3, i+1))
                i=i+1
                subplot_title=("Subplot: "+str(key))
                axes[-1].set_title(subplot_title)  
                plt.imshow(self.dict[key])
            fig.tight_layout()    
        else:
            plt.imshow(self.dict[which], interpolation='nearest')
        plt.show()
    def NormalizeData(self, data):
        return (data - np.min(data)) / (np.max(data) - np.min(data))

    def ToImage(self):
        i_a_h, j_a_h = np.where(self.human_awards == 1) #indexes of the human_awards
        i_a_c, j_a_c = np.where(self.computer_awards == 1) #indexes of the computer_awards
        
        r = self.board/10 + self.all_awards
        r += self.human_trace
        
        if(not np.any(self.all_awards)):
            g = np.zeros([10,10])
        else:    
            g = self.board + self.all_awards*3
        
        b = self.board/10 + self.all_awards/10
        b += self.computer_trace
        
        if i_a_h.size != 0:
            r[i_a_h, j_a_h] += 0.5
            g[i_a_h, j_a_h] += 0.2
            b[i_a_h, j_a_h] += 0.2
        if i_a_c.size != 0:
            r[i_a_c, j_a_c] += 0.2
            g[i_a_c, j_a_c] += 0.2
            b[i_a_c, j_a_c] += 0.5
        
        r = self.NormalizeData(r)
        g = self.NormalizeData(g)
        b = self.NormalizeData(b)

        rgb = np.dstack((r,g,b))
        return rgb
        
```
